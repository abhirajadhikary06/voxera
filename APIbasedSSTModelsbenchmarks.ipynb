```python
import requests
import time
import re
import os
from jiwer import wer, cer

SPEECHMATICS_API_KEY = "your-env"

pre_text = """title the intelligent voice test passage hello and welcome to the advanced text to speech evaluation test this passage is designed to evaluate pronunciation accuracy pacing tone clarity emotional expression and natural conversational flow in modern ai speech synthesis systems let us begin with a simple and calm sentence the quick brown fox jumps over the lazy dog near the river at sunrise now we introduce numbers and measurements the temperature today is 27 5 degrees celsius and the humidity is 68 percent the server processed 1024 requests in 3 75 seconds with a success rate of 99 94 percent next we test dates time and currency the meeting is scheduled for monday march 23rd 2026 at 10 30 am the project budget is 5000 us dollars or approximately 4 2 lakh indian rupees now let us evaluate abbreviations and technical words the api response was returned in json format using http over tls the devops team deployed the docker container to a kubernetes cluster on aws here comes a sentence with punctuation and pauses wait did you hear that something is not right the system however continues to operate normally now we test emotional variation happy tone i am really excited to see how beautifully this ai voice can speak serious tone this system must be reliable accurate and production ready questioning tone can this voice adapt to different accents and speaking styles warning tone please be careful while deploying this model in real time applications now we test conversational dialogue person a hello how are you today person b i am doing well thank you how about you person a i am testing a new ai speech model for a production system next we include complex technical vocabulary the microservices architecture uses asynchronous communication with message queues load balancers and distributed logging systems machine learning models require proper monitoring latency optimization and real time inference pipelines now we test special characters and symbols email test user example com website https example ai version v2 1 4 beta file path home user projects tts model app py let us test multilingual capability namaste aap kaise hain bonjour comment allez vous hola cómo estás thank you and have a wonderful day finally we end with a natural closing sentence this concludes the intelligent voice testing passage thank you for listening and we hope this test helps you build better and more natural ai speech systems"""

def normalize_text(text):
    text = text.lower()
    text = re.sub(r'https\S+', ' ', text)
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def test_speechmatics_batch(audio_path, reference_text):
    try:
        print("--- Full Result: Speechmatics (Batch API - Faster) ---")
        headers = {"Authorization": f"Bearer {SPEECHMATICS_API_KEY}"}

        # Step 1: Submit the job
        with open(audio_path, 'rb') as f:
            files = {
                'data_file': f,
                'config': (None, '{"type": "transcription", "transcription_config": {"operating_point": "enhanced", "language": "en"}}')
            }
            start = time.time()
            response = requests.post("https://asr.api.speechmatics.com/v2/jobs", headers=headers, files=files)
            response.raise_for_status()
            job_id = response.json()["id"]

        # Step 2: Poll for completion
        while True:
            job_response = requests.get(f"https://asr.api.speechmatics.com/v2/jobs/{job_id}", headers=headers)
            job_response.raise_for_status()
            status = job_response.json()["job"]["status"]
            if status == "done":
                break
            elif status == "rejected":
                print("Job rejected.")
                return
            time.sleep(2)

        # Step 3: Get transcript
        transcript_response = requests.get(f"https://asr.api.speechmatics.com/v2/jobs/{job_id}/transcript?format=txt", headers=headers)
        transcript_response.raise_for_status()
        full_transcript = transcript_response.text.strip()
        end = time.time()

        norm_pre = normalize_text(reference_text)
        norm_trans = normalize_text(full_transcript)

        word_err = wer(norm_pre, norm_trans)
        char_err = cer(norm_pre, norm_trans)

        print(f"Latency: {end - start:.2f}s | WER: {word_err:.2%} | CER: {char_err:.2%}")
        print(f"Full Transcript:\n{full_transcript}")
        print("-" * 50)

    except Exception as e:
        print(f"Exception during Speechmatics test: {e}")

local_file = "test_audio_16k.wav"
if os.path.exists(local_file):
    test_speechmatics_batch(local_file, pre_text)
else:
    print("Audio file missing for Speechmatics test.")
```

```python
import azure.cognitiveservices.speech as speechsdk
import time
import re
import os
from jiwer import wer, cer

# Replace with your actual credentials
AZURE_KEY    = "your-env"
AZURE_REGION = "your-env"

# The actual reference text provided by the user
pre_text = """title the intelligent voice test passage hello and welcome to the advanced text to speech evaluation test this passage is designed to evaluate pronunciation accuracy pacing tone clarity emotional expression and natural conversational flow in modern ai speech synthesis systems let us begin with a simple and calm sentence the quick brown fox jumps over the lazy dog near the river at sunrise now we introduce numbers and measurements the temperature today is 27 5 degrees celsius and the humidity is 68 percent the server processed 1024 requests in 3 75 seconds with a success rate of 99 94 percent next we test dates time and currency the meeting is scheduled for monday march 23rd 2026 at 10 30 am the project budget is 5000 us dollars or approximately 4 2 lakh indian rupees now let us evaluate abbreviations and technical words the api response was returned in json format using http over tls the devops team deployed the docker container to a kubernetes cluster on aws here comes a sentence with punctuation and pauses wait did you hear that something is not right the system however continues to operate normally now we test emotional variation happy tone i am really excited to see how beautifully this ai voice can speak serious tone this system must be reliable accurate and production ready questioning tone can this voice adapt to different accents and speaking styles warning tone please be careful while deploying this model in real time applications now we test conversational dialogue person a hello how are you today person b i am doing well thank you how about you person a i am testing a new ai speech model for a production system next we include complex technical vocabulary the microservices architecture uses asynchronous communication with message queues load balancers and distributed logging systems machine learning models require proper monitoring latency optimization and real time inference pipelines now we test special characters and symbols email test user example com website https example ai version v2 1 4 beta file path home user projects tts model app py let us test multilingual capability namaste aap kaise hain bonjour comment allez vous hola cómo estás thank you and have a wonderful day finally we end with a natural closing sentence this concludes the intelligent voice testing passage thank you for listening and we hope this test helps you build better and more natural ai speech systems"""

def normalize_text(text):
    """Lowercases text and removes punctuation for fairer WER/CER scoring."""
    text = text.lower()
    text = re.sub(r'https\S+', ' ', text)
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def test_azure_speech(audio_path, reference_text):
    try:
        print("--- Full Result: Microsoft Azure Speech (Detailed API) ---")
        speech_config = speechsdk.SpeechConfig(subscription=AZURE_KEY, region=AZURE_REGION)

        # Betterment options: Detailed output for N-best and confidence, set profanity to Raw
        speech_config.output_format = speechsdk.OutputFormat.Detailed
        speech_config.set_profanity(speechsdk.ProfanityOption.Raw)

        audio_config  = speechsdk.audio.AudioConfig(filename=audio_path)
        recognizer    = speechsdk.SpeechRecognizer(speech_config=speech_config, audio_config=audio_config)

        start = time.time()
        # recognize_once is faster for short files; for longer use start_continuous_recognition
        result = recognizer.recognize_once()
        end = time.time()

        if result.reason == speechsdk.ResultReason.RecognizedSpeech:
            transcript = result.text
            norm_pre = normalize_text(reference_text)
            norm_trans = normalize_text(transcript)

            word_err = wer(norm_pre, norm_trans)
            char_err = cer(norm_pre, norm_trans)

            print(f"Latency: {end - start:.2f}s | WER: {word_err:.2%} | CER: {char_err:.2%}")
            print(f"Full Transcript:\n{transcript}")
        else:
            print(f"Speech Recognition Failed: {result.reason}")
            if result.reason == speechsdk.ResultReason.Canceled:
                print(f"Cancellation details: {result.cancellation_details.error_details}")
        print("-" * 50)

    except Exception as e:
        print(f"Exception during Azure test: {e}")

local_file = "test_audio_16k.wav"
if os.path.exists(local_file) and AZURE_KEY != "YOUR_AZURE_SPEECH_KEY":
    test_azure_speech(local_file, pre_text)
else:
    print("Audio file missing or Azure credentials not provided.")
```

```python
import boto3, time, uuid, json, re, os
from jiwer import wer, cer
import urllib.request

# Replace with your actual credentials
AWS_ACCESS_KEY = "YOUR_AWS_ACCESS_KEY"
AWS_SECRET_KEY = "YOUR_AWS_SECRET_KEY"
AWS_REGION     = "us-east-1"
S3_BUCKET      = "YOUR_S3_BUCKET_NAME"

# The actual reference text provided by the user
pre_text = """title the intelligent voice test passage hello and welcome to the advanced text to speech evaluation test this passage is designed to evaluate pronunciation accuracy pacing tone clarity emotional expression and natural conversational flow in modern ai speech synthesis systems let us begin with a simple and calm sentence the quick brown fox jumps over the lazy dog near the river at sunrise now we introduce numbers and measurements the temperature today is 27 5 degrees celsius and the humidity is 68 percent the server processed 1024 requests in 3 75 seconds with a success rate of 99 94 percent next we test dates time and currency the meeting is scheduled for monday march 23rd 2026 at 10 30 am the project budget is 5000 us dollars or approximately 4 2 lakh indian rupees now let us evaluate abbreviations and technical words the api response was returned in json format using http over tls the devops team deployed the docker container to a kubernetes cluster on aws here comes a sentence with punctuation and pauses wait did you hear that something is not right the system however continues to operate normally now we test emotional variation happy tone i am really excited to see how beautifully this ai voice can speak serious tone this system must be reliable accurate and production ready questioning tone can this voice adapt to different accents and speaking styles warning tone please be careful while deploying this model in real time applications now we test conversational dialogue person a hello how are you today person b i am doing well thank you how about you person a i am testing a new ai speech model for a production system next we include complex technical vocabulary the microservices architecture uses asynchronous communication with message queues load balancers and distributed logging systems machine learning models require proper monitoring latency optimization and real time inference pipelines now we test special characters and symbols email test user example com website https example ai version v2 1 4 beta file path home user projects tts model app py let us test multilingual capability namaste aap kaise hain bonjour comment allez vous hola cómo estás thank you and have a wonderful day finally we end with a natural closing sentence this concludes the intelligent voice testing passage thank you for listening and we hope this test helps you build better and more natural ai speech systems"""

def normalize_text(text):
    """Lowercases text and removes punctuation for fairer WER/CER scoring."""
    text = text.lower()
    text = re.sub(r'https\S+', ' ', text)
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def test_amazon_transcribe(audio_path, reference_text):
    if AWS_ACCESS_KEY == "YOUR_AWS_ACCESS_KEY":
        print("AWS credentials not provided.")
        return

    try:
        print("--- Full Result: Amazon Transcribe (S3 Upload) ---")
        s3_client = boto3.client("s3", aws_access_key_id=AWS_ACCESS_KEY,
                                  aws_secret_access_key=AWS_SECRET_KEY, region_name=AWS_REGION)
        transcribe_client = boto3.client("transcribe", aws_access_key_id=AWS_ACCESS_KEY,
                                          aws_secret_access_key=AWS_SECRET_KEY, region_name=AWS_REGION)

        s3_key = f"test_audio_{uuid.uuid4().hex}.wav"
        s3_client.upload_file(audio_path, S3_BUCKET, s3_key)
        s3_uri = f"s3://{S3_BUCKET}/{s3_key}"

        job_name = f"voxera-test-{uuid.uuid4().hex}"

        start = time.time()
        # Betterment: enable automatic punctuation and use standard model
        transcribe_client.start_transcription_job(
            TranscriptionJobName=job_name,
            Media={"MediaFileUri": s3_uri},
            MediaFormat="wav",
            LanguageCode="en-US",
            Settings={'ShowAlternatives': False}
        )

        while True:
            status = transcribe_client.get_transcription_job(TranscriptionJobName=job_name)
            state  = status["TranscriptionJob"]["TranscriptionJobStatus"]
            if state in ["COMPLETED", "FAILED"]:
                break
            time.sleep(3)

        if state == "FAILED":
            print(f"Transcription job failed: {status['TranscriptionJob']['FailureReason']}")
            return

        transcript_uri = status["TranscriptionJob"]["Transcript"]["TranscriptFileUri"]
        with urllib.request.urlopen(transcript_uri) as response:
            data = json.loads(response.read())

        transcript = data["results"]["transcripts"][0]["transcript"].strip()
        end = time.time()

        norm_pre = normalize_text(reference_text)
        norm_trans = normalize_text(transcript)

        word_err = wer(norm_pre, norm_trans)
        char_err = cer(norm_pre, norm_trans)

        print(f"Latency: {end - start:.2f}s | WER: {word_err:.2%} | CER: {char_err:.2%}")
        print(f"Full Transcript:\n{transcript}")
        print("-" * 50)

    except Exception as e:
        print(f"Exception during Amazon Transcribe test: {e}")

local_file = "test_audio_16k.wav"
if os.path.exists(local_file):
    test_amazon_transcribe(local_file, pre_text)
else:
    print("Audio file missing for Amazon Transcribe.")
```
