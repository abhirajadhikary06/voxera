## Upload & Standardize Audio

```python
import time
import os
from google.colab import files
from pydub import AudioSegment

print("Please upload your audio file (mp3, wav, m4a, etc.):")
uploaded = files.upload()
original_filename = list(uploaded.keys())[0]

normalized_audio_path = "test_audio_16k.wav"

# Convert audio to 16kHz, 16-bit, Mono WAV
print("Normalizing audio format...")
audio = AudioSegment.from_file(original_filename)
audio = audio.set_frame_rate(16000).set_channels(1).set_sample_width(2)
audio.export(normalized_audio_path, format="wav")
print(f"Audio ready: {normalized_audio_path}")
```

### Pocketsphinx

```python
from pocketsphinx import AudioFile
import time
from jiwer import wer, cer

print("--- 1. Pocketsphinx ---")
start_time = time.time()

# Process audio
config = {
    'audio_file': normalized_audio_path,
}
text = []
for phrase in AudioFile(**config):
    text.append(str(phrase))

process_time = time.time() - start_time
transcript = ' '.join(text)

print(f"Transcript: {transcript}")
print(f"Time Taken: {process_time:.2f} seconds\n")

# Evaluation
prediction_text = transcript
wer_score = wer(ground_truth_text, prediction_text)
cer_score = cer(ground_truth_text, prediction_text)

print("Evaluation Metrics:")
print(f"WER: {wer_score:.4f}")
print(f"CER: {cer_score:.4f}")
```

```python
import speech_recognition as sr
import time
from jiwer import wer, cer

print("--- 2. SpeechRecognition (Google API) ---")
recognizer = sr.Recognizer()

with sr.AudioFile(normalized_audio_path) as source:
    audio_data = recognizer.record(source)

start_time = time.time()
try:
    text = recognizer.recognize_google(audio_data)
    process_time = time.time() - start_time

    print(f"Transcript: {text}")
    print(f"Time Taken: {process_time:.2f} seconds\n")

    # Evaluation
    prediction_text = text
    wer_score = wer(ground_truth_text, prediction_text)
    cer_score = cer(ground_truth_text, prediction_text)

    print("Evaluation Metrics:")
    print(f"WER: {wer_score:.4f}")
    print(f"CER: {cer_score:.4f}\n")

except sr.UnknownValueError:
    print("SpeechRecognition could not understand audio")
except sr.RequestError as e:
    print(f"Could not request results; {e}")
```

```python
import speech_recognition as sr
import time
from jiwer import wer, cer

print("--- 2. SpeechRecognition (CMU Sphinx - Offline) ---")

recognizer = sr.Recognizer()

with sr.AudioFile(normalized_audio_path) as source:
    audio_data = recognizer.record(source)

start_time = time.time()

try:
    # CMU Sphinx offline recognition
    text = recognizer.recognize_sphinx(audio_data)
    process_time = time.time() - start_time

    transcript = text.strip()

    print(f"Transcript: {transcript}")
    print(f"Time Taken: {process_time:.2f} seconds\n")

    # -------------------------
    # JiWER Evaluation
    # -------------------------
    prediction_text = transcript
    wer_score = wer(ground_truth_text, prediction_text)
    cer_score = cer(ground_truth_text, prediction_text)

    print("Evaluation Metrics:")
    print(f"WER: {wer_score:.4f}")
    print(f"CER: {cer_score:.4f}\n")

except sr.UnknownValueError:
    print("CMU Sphinx could not understand audio")

except sr.RequestError as e:
    print(f"Sphinx error: {e}")
```

### Vosk

```python
# Download lightweight english model
!wget -q https://alphacephei.com/vosk/models/vosk-model-small-en-us-0.15.zip
!unzip -q vosk-model-small-en-us-0.15.zip
```

```python
from vosk import Model, KaldiRecognizer
import wave
import json
import time
from jiwer import wer, cer

print("--- 3. Vosk ---")
model = Model("vosk-model-small-en-us-0.15")
wf = wave.open(normalized_audio_path, "rb")
rec = KaldiRecognizer(model, wf.getframerate())

start_time = time.time()
results = []
while True:
    data = wf.readframes(4000)
    if len(data) == 0:
        break
    if rec.AcceptWaveform(data):
        part_result = json.loads(rec.Result())
        results.append(part_result.get("text", ""))

final_result = json.loads(rec.FinalResult())
results.append(final_result.get("text", ""))
process_time = time.time() - start_time

transcript = ' '.join(results)

print(f"Transcript: {transcript}")
print(f"Time Taken: {process_time:.2f} seconds\n")

# Evaluation
prediction_text = transcript
wer_score = wer(ground_truth_text, prediction_text)
cer_score = cer(ground_truth_text, prediction_text)

print("Evaluation Metrics:")
print(f"WER: {wer_score:.4f}")
print(f"CER: {cer_score:.4f}\n")
```

### WhisperX

```python
import whisperx
import time

print("--- 2. WhisperX ---")

# Setup device (change to "cuda" if using a Colab GPU!)
device = "cpu"
batch_size = 8
compute_type = "int8"  # Change to "float16" if using GPU

# Load model and audio
model = whisperx.load_model("base", device, compute_type=compute_type)
audio = whisperx.load_audio(normalized_audio_path)

start_time = time.time()

# Transcribe
result = model.transcribe(audio, batch_size=batch_size)

# Combine chunks into a single string
text = " ".join([segment["text"] for segment in result["segments"]])

process_time = time.time() - start_time
transcript = text.strip()

print(f"Transcript: {transcript}")
print(f"Time Taken: {process_time:.2f} seconds\n")

# -------------------------
# JiWER Evaluation (after execution)
# -------------------------
from jiwer import wer, cer

reference_text = ground_truth_text  # original sentence
prediction_text = transcript

wer_score = wer(reference_text, prediction_text)
cer_score = cer(reference_text, prediction_text)

print("Evaluation Metrics:")
print(f"WER: {wer_score:.4f}")
print(f"CER: {cer_score:.4f}\n")
```
